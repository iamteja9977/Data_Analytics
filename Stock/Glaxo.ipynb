# Repeat the same 4 operation for GLAXO file , as BEML 

```python
#1
import pandas as pd
df=pd.read_csv('GLAXO.csv')
df.head(5)
```

```python
#2
df2 = df[['Date', "Close"]]
df2.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline  
plt.plot(df2.Close)
plt.xlabel("Time")
plt.ylabel("Close Price")
plt.show()
```

```python
#3
df2 = df2.set_index(pd.DatetimeIndex(df2['Date']))
df2.head(5)
```

```python

#4
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline  
plt.plot(df2.Close)
plt.xlabel("Time")
plt.ylabel("Close Price")
plt.show()
```

```python
destination_file = "NewGLAXO.csv"
df2.to_csv(destination_file, index=False)
print("ETL process completed.")
```
